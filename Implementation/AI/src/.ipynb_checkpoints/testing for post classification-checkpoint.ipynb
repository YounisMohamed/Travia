```python
from ultralytics import YOLO

class YoloDescriber:
    def __init__(self, model_path="yolov8n.pt"):
        self.yolo_model = YOLO(model_path)

    def get_yolo_description(self, image_path):
        """Generate natural language description from YOLO detections"""

        # Run YOLO detection
        results = self.yolo_model(image_path, verbose=False)

        detected_objects = []

        # Extract high-confidence objects
        for result in results:
            boxes = result.boxes
            if boxes is not None:
                for box in boxes:
                    class_id = int(box.cls[0])
                    confidence = float(box.conf[0])
                    class_name = self.yolo_model.names[class_id]

                    if confidence > 0.0:  # Only confident detections
                        detected_objects.append(class_name)

        # Convert objects to natural description
        if not detected_objects:
            return ""

        # Group similar objects
        object_counts = {}
        for obj in detected_objects:
            object_counts[obj] = object_counts.get(obj, 0) + 1

        # Create natural language description
        descriptions = []
        for obj, count in object_counts.items():
            if count == 1:
                descriptions.append(f"a {obj}")
            else:
                descriptions.append(f"{count} {obj}s")

        # Join naturally
        if len(descriptions) == 1:
            yolo_description = f"contains {descriptions[0]}"
        elif len(descriptions) == 2:
            yolo_description = f"contains {descriptions[0]} and {descriptions[1]}"
        else:
            yolo_description = f"contains {', '.join(descriptions[:-1])}, and {descriptions[-1]}"

        return yolo_description
```

```python
def load_blip_model():
    vision_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    vision_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
    return vision_processor, vision_model

def get_blip_description(vision_processor, vision_model, image_path):
    """Generate a natural language caption from an image using BLIP"""
    image = Image.open(image_path).convert('RGB')
    inputs = vision_processor(image, return_tensors="pt")
    output = vision_model.generate(**inputs, max_length=50)
    blip_caption = vision_processor.decode(output[0], skip_special_tokens=True)
    return blip_caption
```

# map comined text to attributes LLM approach

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import json
from typing import List, Dict, Optional
import re

# Approach 1: Using Small Local LLMs (Best Overall)
class LLMClassifier:
    """
    Uses instruction-tuned LLMs for classification
    Accuracy: 95-98% without any training
    """
    
    def __init__(self, model_name: str = "microsoft/Phi-3-mini-4k-instruct"):
        """
        Initialize with a small but powerful model
        Options:
        - "microsoft/Phi-3-mini-4k-instruct" (3.8B params, very fast)
        - "mistralai/Mistral-7B-Instruct-v0.2" (7B params, more accurate)
        - "google/flan-t5-xl" (3B params, good for classification)
        """
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForCausalLM.from_pretrained(
            model_name,
            torch_dtype=torch.float16,
            device_map="auto"
        )
        
    def create_prompt(self, caption: str, blip_text: str, yolo_objects: str) -> str:
        """Create a structured prompt for the LLM"""
        prompt = f"""You are a restaurant/venue classifier. Analyze this place based on the given information.

                    User Caption: {caption}
                    Visual Description: {blip_text}
                    Objects Detected: {yolo_objects}

                    Classify this place according to these MUTUALLY EXCLUSIVE pairs:
                    1. Either "romantic" OR "good_for_kids" (not both)
                    2. Either "classy" OR "casual" (not both)

                    Consider:
                    - Romantic: intimate, couples, date night, candles, wine, cozy
                    - Good_for_kids: family-friendly, children welcome, kid menus, high chairs
                    - Classy: upscale, elegant, fine dining, sophisticated, formal
                    - Casual: relaxed, informal, laid-back, simple, everyday dining

                    Output ONLY a JSON object with exactly this format:
                    {{"romantic": true/false, "good_for_kids": true/false, "classy": true/false, "casual": true/false}}"""

        return prompt

    
    def classify(self, caption: str, blip_text: str, yolo_objects: str) -> Dict:
        """Classify using LLM reasoning"""
        prompt = self.create_prompt(caption, blip_text, yolo_objects)

        # Generate response
        inputs = self.tokenizer(prompt, return_tensors="pt")

        with torch.no_grad():
            outputs = self.model.generate(
                **inputs,
                max_new_tokens=50,
                temperature=0.1,
                do_sample=True
            )

        response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Extract JSON from response
       # Extract JSON from response
        try:
            json_match = re.search(r'\{.*\}', response[len(prompt):])
            if json_match:
                result = json.loads(json_match.group())

                # Enforce mutual exclusivity
                if result.get('romantic') and result.get('good_for_kids'):
                    if 'romantic' in caption.lower() or 'date' in caption.lower():
                        result['good_for_kids'] = False
                    else:
                        result['romantic'] = False

                if result.get('classy') and result.get('casual'):
                    if 'elegant' in blip_text.lower() or 'upscale' in caption.lower():
                        result['casual'] = False
                    else:
                        result['classy'] = False

                return {
                    'attributes': result,
                    'model_response': response[len(prompt):].strip(),
                    'confidence': 'high'
                }

            # ❗ ADD THIS: if json_match failed, fallback explicitly
            return self._fallback_classification(caption, blip_text, yolo_objects)

        except:
            return self._fallback_classification(caption, blip_text, yolo_objects)


    
    def _fallback_classification(self, caption: str, blip_text: str, yolo_objects: str) -> Dict:
        """Simple fallback if LLM fails"""
        combined = f"{caption} {blip_text} {yolo_objects}".lower()

        return {
            'attributes': {
                'romantic': 'romantic' in combined or 'date' in combined,
                'good_for_kids': 'kid' in combined or 'family' in combined,
                'classy': 'elegant' in combined or 'upscale' in combined,
                'casual': 'casual' in combined or 'relaxed' in combined
            },
            'confidence': 'low'
        }
```

# map comined text to attributes ZERO SHOT CLASSIFIER

```python
from transformers import pipeline

# Load the zero-shot classification pipeline
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=0  # set to -1 for CPU, or 0 for CUDA/GPU if available
)

def zero_shot_classify(caption: str, blip_text: str, yolo_objects: str) -> dict:
    # Combine all text sources for context
    combined_text = f"{caption} {blip_text} {yolo_objects}"

    # The mutually exclusive labels you want
    labels = ["romantic", "good_for_kids", "classy", "casual"]

    # Run zero-shot classification
    result = classifier(combined_text, labels, multi_label=True)

    # Build a dict of label -> confidence (threshold 0.5 can be adjusted)
    attributes = {label: score > 0.5 for label, score in zip(result['labels'], result['scores'])}

    return {
        'attributes': attributes,
        'confidence_scores': dict(zip(result['labels'], result['scores'])),
        'raw_output': result
    }
```

```python
from transformers import AutoProcessor, AutoModelForImageClassification
from PIL import Image
import torch

# Load the model and processor
model = AutoModelForImageClassification.from_pretrained("Kaludi/food-category-classification-v2.0")
processor = AutoProcessor.from_pretrained("Kaludi/food-category-classification-v2.0")

# Load your image
path = r"C:\Users\AbdulRahman\Desktop\aa.jpg"
image = Image.open(path).convert("RGB")

# Process and predict
inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)
predictions = torch.nn.functional.softmax(outputs.logits, dim=1)

# Get top prediction
label_id = predictions.argmax().item()
label = model.config.id2label[label_id]
confidence = predictions[0, label_id].item()

print(f"Predicted Cuisine: {label} with confidence {confidence:.2f}")
```
